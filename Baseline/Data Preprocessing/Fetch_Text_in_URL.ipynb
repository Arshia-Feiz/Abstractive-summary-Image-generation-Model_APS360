```python
from newspaper import Article
import nltk

# Download the punkt tokenizer once
nltk.download('punkt')

def extract_news_text(url):
    try:
        article = Article(url)
        article.download()
        article.parse()
        #article.nlp()  # Optional, for summary/keywords

        print("Title:", article.title)
        print("\nText:\n", article.text)
        #print("\nSummary:\n", article.summary)
    except Exception as e:
        print("Failed to extract article:", e)

if __name__ == "__main__":
    url = input("Enter news article URL: ")
    extract_news_text(url)
```
